```python
import matplotlib.pyplot as plt
import json
```

```python
with open("kmc/build/out.json", "r+") as f:
    result = json.load(f)
```

```python
result.keys()
```

```python
amounts = result['amounts']
amount_traces = [[a[i] for a in amounts] for i in range(len(amounts[0]))]
```

```python
for idx, s_name in enumerate(result['specieNames']):
    plt.plot(result['times'], amount_traces[idx], label=s_name)
```
